```python
import phthalicacid as pa

smiles, iso, cas = pa.samples_data()
```

```python
test_data = smiles[1:25]
iso_data = iso[1:25]
cas1 = []
samples1 = []
for i in range(len(smiles)):
    test = pa.select_smiles(smiles[i],iso[i])
    if test:
        samples1.append(test)
        cas1.append(cas[i])
print(len(samples1))
print(cas1)
print(len(cas1))
```

```python
detects = []
cas2 = []
samples2 = []
for i in range(len(samples1)):
    dicarbo = pa.detect_phthalic_acid(samples1[i])
    if dicarbo:
        detects.append(dicarbo)
        cas2.append(cas1[i])
        samples2.append(samples1[i])
print(len(detects))
print(len(cas2))
```

```python
coords = []
coords_ion = []
cas3 = []
for i in range(len(samples2)):
    coord, coord_ion = pa.generate_coord(samples2[i], detects[i])
    if coord and coord_ion:
        coords.append(coord)
        coords_ion.append(coord_ion)
        cas3.append(cas2[i])
print(len(coords),len(coords_ion))
print(len(cas3))
```
