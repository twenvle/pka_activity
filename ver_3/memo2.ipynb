```python
from rdkit import Chem
from rdkit.Chem import AllChem

smiles1 = "O=C(O)C1=CC=C(C=C1C(=O)O)N(=O)=O"
smiles2 = "O=C(C1=CC(C(O)=O)=C(C(O)=O)C=C1C(O)=O)O"
smiles3 = "FC1=C(F)C(F)=C(C(O)=O)C(C(O)=O)=C1F"

def get_idx(smiles):
    mol = Chem.MolFromSmiles(smiles)
    mol = Chem.AddHs(mol)  # 水素を追加
    phthalic_pattern = Chem.MolFromSmarts("c1ccc(C(=O)O)c(C(=O)O)c1")
    
    # フタル酸構造を持つかどうか
    matches_data = mol.GetSubstructMatches(phthalic_pattern)
    print(matches_data)
    atoms = ""
    for match in matches_data[0]:
        atom = mol.GetAtomWithIdx(match)  # その番号の原子の情報
        symbol = atom.GetSymbol()
        neighbors = atom.GetNeighbors()
        for neighbor in neighbors:
            idx = neighbor.GetIdx()
            # print(idx)
        atoms += f"{match}: {symbol}, "
    print(atoms)
    print(type(matches_data[0][0]))

get_idx(smiles1)
get_idx(smiles2)
get_idx(smiles3)
```
